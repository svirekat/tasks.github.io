## индивидуальная задача 2 (вариант 9)

Дано генеалогическое древо, определите высоту всех его элементов. Каждом элементу дерева сопоставляется целое неотрицательное число, называемое высотой. У родоначальника высота равна 0, у любого другого элемента высота на 1 больше, чем у его родителя.

Программа получает на вход число элементов в генеалогическом древе N. Далее следует N строк, задающие родителя для каждого элемента древа, кроме родоначальника. Каждая строка имеет вид имя_потомка имя_родителя.
Программа должна вывести список всех элементов древа в лексикографическом порядке. После вывода имени каждого элемента необходимо вывести его высоту.


```python
with open('text_ind2.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()
    N = int(lines[0])
    persons_highs = {}
    for i in range(1, N+1):
        child_parent = lines[i].split()
        if len(child_parent) == 1:
            oldest = child_parent[0]
            persons_highs[oldest] = 0
        else:
            child, parent = child_parent[0], child_parent[1]
            persons_highs[child] = persons_highs.get(parent, 0) + 1
f.close()
sorted_map = sorted(persons_highs)
for i in range(N):
    print(sorted_map[i] + ", " + str(persons_highs.get(sorted_map[i])))
```
