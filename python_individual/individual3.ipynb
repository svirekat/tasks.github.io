## индивидуальная задача 3 (вариант 1)

Банк разрабатывает новый тип банковских карт с идентификацией клиента. Каждая карта имеет кодовое число X, из которого можно получить номер счета клиента, стирая некоторые цифры.

Необходимо определить минимальный номер  счета N, который нельзя получить из цифр X указанным способом. Например, при X = 12013456789 номера счетов 5, 12, 17 или 12013456789 получить можно, а номера 22 или 71 получить нельзя.

Напишите программу, которая найдет значение N по заданному X.
На вход программе подается натуральное число X без ведущих нулей. Выведите искомое значение N без ведущих нулей.

Ввод
179 
Вывод
2

Ввод
12013456789 
Вывод
22

```python
import re
X = input()
N = 1
while True:
    match = re.search('.*'.join(d for d in str(N)), X)
    if match:
        N = N + 1
    else:
        break
print(N)
```
