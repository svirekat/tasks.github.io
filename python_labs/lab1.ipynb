## Задачи без использования ветвления, циклов, функций

1. Дано натуральное число n. Определите, сколько часов и минут
будут показывать электронные часы. Введите целое неотрицательное число n.
Выведите количество часов (0-23) и количество минут (0-59).

2. Дано n школьников и k яблок. Нужно разделить яблоки "почти
поровну", так чтобы разница в количестве яблок между любыми двумя
школьниками была не больше 1. Необходимо определить, сколько школьников
получат меньше яблок, чем другие школьники.


```python
n = int(input("введите n (кол-во школьников):"))
k = int(input("введите k (кол-во яблок):"))
print("ответ:", n - k%n)
```

3. Дано натуральное число n. Выведите следующее за ним чётное
число.

```python
n = int(input("введите n:"))
print(n + 2 - n%2)
```

4. Сколько дней будет длится маршрут длиной m километров, если
машина каждый день проезжает n километров?

```python
n = int(input("введите n:"))
m = int(input("введите m:"))
print(m // n + (m%n * 1))
```

5. Дано два натуральных числа a и b. Найти наибольшее значение
из них.

```python
a = int(input("введите a:"))
b = int(input("введите b:"))
print(a if a > b else b)
```

## Задачи без использования циклов, функций

1. Дано натуральное число N. Сколько разрезов необходимо сделать,
чтобы разделить пиццу на N одинаковых по форме и размеру кусков?
Например, для N=6 потребуется 3 разреза, а для N=5 - 5 разрезов.

```python
N = int(input("введите N:"))
if N % 2 == 0:
    print(N // 2)
else:
    print(N)
```

2. Дано два натуральных числа N и M. Петя стоит на поле размером
N×М метров на расстоянии x метров от одной из длинных сторон
(не обязательно от ближайшей) и y метров от одной из коротких сторон. Какое
минимальное расстояние должен пройти Петя, чтобы дойти до края поля?
Изначально неизвестно, какая сторона является длинной.

```python
N = int(input("введите N:"))
M = int(input("введите M:"))
x = 4
y = 7
if (N > M):
    if (M-x < x):
        x = M-x
    if (N-y < y):
        y = N-y
else:
    if (N-x < x):
        x = N-x
    if (M-y < y):
        y = M-y
print(x if x < y else y)
    
```

## Задачи без использования функций

1. Дано два натуральных числа A и B (A⩽B) вывести все чётные
числа на отрезке от A до B (Без использования условного оператора if).

```python
A = int(input("введите A:"))
B = int(input("введите B:"))
start = A + (A % 2)
for x in range(start, B+1, 2):
    print(x)
```

2. Дано два натуральных числа a и b. Найти все целые числа от a
до b, которые имеют остаток c при делении на d (Без использования
операторов сравнения, условного оператора и с использованием только
одного цикла).


```python
a = int(input("введите a:"))
b = int(input("введите b:"))
c = 2
d = 3
start = a + c - (a % d)
for x in range(start, b+1, d):
    print(x)
```

3. Дано натуральное число n. Найти сумму 1!+2!+3!+...+n!. (С
использованием только одного цикла).

```python
n = int(input("введите n:"))
s = 0
f = 1
for i in range(1, n+1):
    f *= i
    s += f
print(s)
```
