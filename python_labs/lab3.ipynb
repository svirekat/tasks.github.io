4. Необходимо узнать, какую оценку получит ученик в четверти. Учитель
придерживается следующей системы: вычисляется среднее арифметическое всех оценок в
журнале, и ставится ближайшая целая оценка, не превосходящая среднего
арифметического.
При этом если у школьника есть двойка, а следующая за ней оценка – не двойка, то
двойка считается закрытой, и при вычислении среднего арифметического не учитывается.

```python
grades = list(map(int, input().split()))
s = 0
k = 0
for i in range(len(grades)-1):
    if grades[i] == 2 and grades[i+1] > 2:
        continue
    s += grades[i]
    k += 1
s += grades[-1]
k += 1
print(s // k)
```

8. Дан список чисел. В списке все элементы различны. Поменяйте местами
минимальный и максимальный элемент этого списка.

```python
list8 = [31, 25, 66, 4, -3, 12, 7, -30, 1, 70]
ind_min = list8.index(min(list8))
ind_max = list8.index(max(list8))
list8[ind_min], list8[ind_max] = list8[ind_max], list8[ind_min]
print(list8)
```

9. Найти наибольшую сумму чисел в подпоследовательности заданной длины.
Дано: список чисел, длина подпоследовательности, для которой нужно найти наибольшую
сумму.

```python
list9 = [3, 2, 5, 4, 3, 2, 4]
d = 3
maxsum = 0
for i in range(len(list9) - d + 1):
    s = sum(list9[i:i+d])
    if s > maxsum:
        maxsum = s
print(maxsum)
```
