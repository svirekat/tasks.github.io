## регулярные выражения

1. Дана строка. Получить первое слово из строки

```python
import re
st = "первое слово из строки"
match = re.search(r'^\b\w+\b', st)
if match:
    print({match.group()})
```

2. Дана строка. Получить последнее слово из строки

```python
match = re.search(r'\b\w+\b$', st)
if match:
    print({match.group()})
```

3. Дана последовательность строк. Нужно вывести те, в которых «кот» встречается в 
качестве отдельного слова.

```python
lines = ["маленький котёнок", "большой котёл", "серый кот мяукнул", "котята спят", "белый кот и чёрный кот"]
for l in lines:
    match = re.search(r'\bкот\b', l)
    if match:
        print(l)
```

4. Вводится последовательность строк. Нужно вывести строки, в которых фрагмент 
«кот» присутствует в качестве подстроки не менее 2 раз.

```python
lines = ["кот кот кот", "коты коты коты коты", "кошка и кот", "котята котики", "крупный рогатый скот", "бойкот бойкот", "котята"]
for l in lines:
    match_list = re.findall(r'кот', l)
    if len(match_list) >= 2:
        print(l)
```

5. В России для регистрации машин используют принятый стандарт регистрационных 
знаков. Для частных автомобилей он состоит из серии, самого номера и кода региона 
регистрации. Запись автомобильного номера выглядит так: С065МК777. На вход 
программы подается список автомобильных номеров. Проверить соответствие их 
записи с государственным стандартом.

```python
nomera = ["C065MK777", "C065", "MK777", "V901LD145", "I123PU445A"]
for nomer in nomera:
    match = re.search(r'^[A-Z]\d{3}[A-Z]{2}\d{3}$', nomer)
    if match:
        print('соответствует')
    else:
        print('не соответствует')
```

6. Проверить формат телефонного номера. Номер должен состоять из 11 знаков и
начинаться с 7 или 8. На вход подается список телефонных номеров. Если номер
корректный, то следует вернуть Correct, иначе — Incorrect.

```python
def is_correct_number(number):
    match = re.search(r'^(\+7|8)[\-]?\d{3}[\-]?\d{3}[\-]?\d{2}[\-]?\d{2}$', number)
    if match:
        return "Correct"
    return "Incorrect"
print(is_correct_number("+7-123-4567890"))
```

7. Дана строка. Вывести слова, состоящие из двух одинаковых слогов.

```python
text = "мама папа дочка дядя тётя племянник"
text = text.split()
for word in text:
    match = re.search(r'^([A-Za-zА-Яа-я]+)\1', word)
    if match:
        print(word)
```

8. Вводится последовательность строк. В каждой строке нужно поменять местами две
первые буквы в каждом слове, состоящем из двух и более букв.

```python
strings = ["cat", "ok", "o", "p", "pig", "cloud", "garden", "d", "1ab23"]
pattern = re.compile(r'^[A-Za-zА-Яа-я]{2,}$') 
strings_new = []
for s in strings:
    if len(s) >= 2:
        s = re.sub(pattern, s[1]+s[0]+s[2:], s)
    strings_new.append(s)
print(strings_new)
```

9. Дана строка. Нужно извлечь все имена и фамилии из текста.


```python
text = 'В прошлое воскресенье мы с друзьями решили устроить пикник в парке \
"Сокольники". Компания подобралась разношёрстная: \
пришли Иван Кузнецов со своей девушкой Екатериной Морозовой, \
его брат Дмитрий с женой Ольгой Власовой-Кузнецовой и их маленький сын Петя.'

match_list = re.findall(r'\b[А-ЯA-Z][а-яa-z]{1,}[\-]?[а-яa-z]{1,}\b', text)
if len(match_list) > 0:
    print(match_list)
```

10. Написать регулярное выражение для извлечения из текста всех email-адресов.

```python
text = 'Я зарегистрировался на вашем сайте с адресом user123@mail.ru, но не получил письмо для \
подтверждения регистрации. Проверил папку "Спам" — там пусто. Мой резервный адрес: my.second.email@gmail.com. \
Пожалуйста, продублируйте письмо туда или на корпоративный адрес ivan.petrov@mail.ru.'

match_list = re.findall(r'\b[\w.-]+@(?:gmail\.com|mail\.ru)\b', text)
if len(match_list) > 0:
    print(match_list)
```
