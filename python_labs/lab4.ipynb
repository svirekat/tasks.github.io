1. Дан список чисел. Определить, сколько в нем встречается различных чисел.
Вывести эти числа

```python
list1 = [i for i in (map(int, input().split()))]
print(list1)
set1 = {i for i in list1}
print(len(set1))
```

2. Даны два списка чисел. Найти, сколько чисел содержится одновременно в первом 
списке и во втором списке. Вывести эти числа в порядке возрастания

```python
list1 = [23, 4, 55, 6, 16, 77]
list2 = [41, 55, 78, 77, 14, 15, 16]
A = set(list1)
B = set(list2)
print(A.intersection(B))
```

3. Во входной строке записана последовательность чисел через пробел. Для каждого 
числа определить, встречалось ли это число ранее в последовательности или нет.

```python
str3 = "1 2 3 2 3 4"
set3 = set()
for el in str3.split():
    if el in set3:
        print("YES")
    else:
        print("NO")
    set3.add(el)
```

4. Дети играют в разноцветные кубики из своих наборов. Они хотят узнать, сколько 
цветов присутствуют в обоих наборах, а сколько у каждого из них уникальных цветов. 
Помогите детям найти ответ на этот вопрос. Вывксти количество цветов и сами цвета

```python
nabor1 = {'red', 'green', 'yellow', 'black'}
nabor2 = {'white', 'pink', 'green', 'orange', 'red'}
print("в обоих наборах: " + str(len(nabor1.union(nabor2))) + " цв.")
print("в первом: " + str(len(nabor1.difference(nabor2))) + " уникальных цв.")
print(nabor1.difference(nabor2))
print("во втором: " + str(len(nabor2.difference(nabor1))) + " уникальных цв.")
print(nabor2.difference(nabor1))
```

5. Есть N школьников, каждый из которых знает определенное количество языков. 
Необходимо определить, какие языки знают все школьники и какие языки знает хотя бы 
один из них.

```python
N = int(input())
languages = {"english", "russian", "spanish", "swedish", "italian", "romanian", "japanese", "chinese", "german"}
students = []
for i in range(N):
    langs = {i for i in languages if i }
    students.append(langs)
print(students)
```
